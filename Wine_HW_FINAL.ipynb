```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Let's take a look at all of our data

```python
from sklearn.datasets import load_wine
wine = load_wine()
#print(wine)

#Need to also load in unknown wine data
from numpy import genfromtxt

unknown_wine = genfromtxt('unknown_wine.csv', delimiter=',')
unknown = np.delete(unknown_wine, (0), axis=0) #deleting the column headers which appear as nan
unknown
#type(unknown)
```

```python
print(wine.target)
```

```python
#Let's look at the shape of the data...how many rows and columns?
X = wine.data
y = wine.target

print(np.shape(X))
print(np.shape(y))
```

```python
print(X)
```

```python
print(wine.feature_names)
print(wine.target)
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=4)
```

```python
##Eventually do scaling here!!

# # scale the training data

# from sklearn.preprocessing import StandardScaler
# scaler = StandardScaler()
# scaler.fit(X_train)

# X_train = scaler.transform(X_train)
# X_test = scaler.transform(X_test)
```

```python
from sklearn.neural_network import MLPClassifier
mlp = MLPClassifier(hidden_layer_sizes=(10, 10, 10), max_iter=1000)
mlp.fit(X_train, y_train)
```

```python
# try makign some predictions using your classifier
predictions = mlp.predict(X_test)

plt.plot(y_test, predictions, 'ro')
plt.plot([0, 2], [0, 2],'k')
```

```python
# and now test your predictions

from sklearn.metrics import classification_report, confusion_matrix

print(confusion_matrix(y_test,predictions))
print(classification_report(y_test,predictions))

# We will use the confusion matrix to determine the accuracy which is measured as the total 
# number of correct predictions divided by the total number of predictions.
```

```python
# Looping through different combinations of layers and nodes.

scores = {}
scores_list = []
results = []
conf = []
BetterWork = []

l_range = range(2, 10, 1) #number of layers  
n_range = range(4, 10, 1) #number of nodes

from sklearn import metrics

for n in n_range:
    keep = [] #creating the list, leaving empty to append later
    
    for l in l_range:
        keep.append(n)
        mlp = MLPClassifier(hidden_layer_sizes=(keep), max_iter=1000, random_state=4)
        mlp.fit(X_train, y_train)
        y_pred = mlp.predict(X_test)
        score = metrics.accuracy_score(y_test,y_pred)
        data = [l, n, score]
        results.append(data)
        
        if score >= 0.90:
            BETTERwork = data
            BetterWork.append(BETTERwork)
            confusion = confusion_matrix(y_test, y_pred)
            confusion = np.asarray(confusion)
            conf.append(confusion)
            print('Layers:', l, 'Nodes:', n, '\n', metrics.classification_report(y_test,y_pred))
```

The classification report also helps us in determining the effectiveness of our classifier by providing some valuable metrics, namely the precision score, recall score, and the f-measure of each class. In these tables, the precision is the ability of out classifier to not label something positive that is actually negative. The recall is the ability of the classifier to find all positive instances correctly. The f-measure is the weighted harmonic mean (score) of the precision and accuracy for each class. From these tables, it appears there are five conbinations that give us the same high level results for precision, recall, and f-measure, including combinations of layers and nodes of (2,5), (2,9), (5,4), (7,9), and (9,9). We will choose one of these as our combinaiton for running our model with the unknown data.

```python
work = np.asarray(results)

layer = work[:,0]
node = work[:,1]
scr = work[:,2]

dirt = pd.DataFrame({'NODES': node, 'LAYERS': layer, 'Z': scr})
dirt_pivoted = dirt.pivot("NODES", "LAYERS", "Z")
ax = sns.heatmap(dirt_pivoted, annot=True)
plt.show()
```

Based on all of the information from above, I will choose to have 7 layers with 9 nodes in each layer. I am choosing this combination above all the others as I believe having more nodes will result in more accurate results compared to a higher number of layers. This is why I chose a combination that has 9 nodes with 7 layers as opposed to 9 nodes and 9 layers. Let's see what happens. 

```python
#Now we need to chose number of layers and nodes and create a new classifier with them

mlp = MLPClassifier(hidden_layer_sizes=(9,9,9,9,9,9,9), max_iter=1000, random_state=4)
mlp.fit(X_train, y_train)
```

```python
#Here we need to make predictions based on the new data set to determine which classes they belong to
#The variable for the new data is 'unknown'

# here we apply the new model built above to 'new' data

#change to wine classes#
#0 = Wine_1, 1=Wine_2, 2=Wine_3
classes = {0:'Wine 1',1:'Wine 2',2:'Wine 3'}

# Make a prediction for the 'new' data 
# predict for the below two random observations
# x_new = unknown #This data will end up being actual rows from the wine dataset
# x_new_scaled = scaler.transform(x_new) #We will also scale this data as we did to the test and train data above

y_predict = mlp.predict(unknown) #make the predictions

for m in range(len(unknown)):
    print('Unknown' ,m+1,':', classes[y_predict[m]])#printing the predicted wine class for each wine number from unknown_wine.csv
```

```python
a = mlp.predict_proba(unknown)#determining the probability of each wine predicted by the model
a = np.around(a,decimals=2)   #simply changing the decimals so the numbers aren't ridicuously long
for m in range(len(a)):
    print('Unknown Wine:', m+1) #printing the unknown wine number from unknown_wine.csv
    print('   1    2    3')   #Just to have and idea of which wine class is associated with the probability
    print(a[m])               #Printing the actual probability of each wine predicted by the model
```

Interestingly, this combination of layers and nodes results in classifying the unknown wines with 3 of class 0 (Wine 1) and 7 of class 1 (Wine 2), no class 2 (Wine 3). The classifier does, however, have a relatively high probability in correctly guessing the class of the unknown wine. Particularly, the probability of guessing the correct wine class when below 0.90, these probabilities are higher than in any other combination of layers and nodes with an accuracy score of 97%, with the lowest probability score at 0.70 as opposed to other combinations having low values in the 0.3 range.
